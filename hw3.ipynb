```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Leibniz Pi formula

```python
def Leibniz(n):
    i, f = 1, 0
    while (i<=n):
        f = f + 1/((((-1)**(i-1))*(2*i-1)))
        i = i+1
    return f

x=np.arange(1,101,1)
y=[Leibniz(n) for n in x]
a=np.pi/4*x/x
plt.figure(figsize=(8,8))
plt.xlabel('n', fontsize = 15)
plt.plot(x,y,label=r'Leibniz  Formula for $\pi$')
plt.plot(x,a,label=r'$\pi/4$')
plt.legend()
```

# Logistic Map

```python
def logistic(r,f):
    g=[]
    for n in range(1,101):
        f=r*f*(1-f)
        g.append(f)
    return(g)

plt.figure(figsize=(15,15))
plt.subplot(231)
plt.plot(logistic(0.5,0.5), label = '$r=0.5, $f_0$=0.5')
plt.legend(loc='lower right', bbox_to_anchor=(1, 0.5))

plt.subplot(232)
plt.plot(logistic(0.95, 0.5), label = '$r=0.95$, $f_0$=0.5')
plt.legend(loc='lower right', bbox_to_anchor=(1, 0.5))

plt.subplot(233)
plt.plot(logistic(2, 0.5), label = '$r=2$, $f_0$=0.5')
plt.legend(loc='lower right', bbox_to_anchor=(1, 0.5))

plt.subplot(234)
plt.plot(logistic(3.2, 0.5), label = '$r=3.2$, $f_0$=0.5')
plt.legend(loc='lower right', bbox_to_anchor=(1, 0.5))

plt.subplot(235)
plt.plot(logistic(3.8, 0.5), label = '$r=3.8$, $f_0$=0.5')
plt.legend(loc='lower right', bbox_to_anchor=(1, 0.5))

plt.subplot(236)
plt.plot(logistic(3.8, 0.501), label = '$r=3.8$, $f_0$=0.501')
plt.legend(loc='lower right', bbox_to_anchor=(1, 0.5))
```

# Sinc function

```python
def facto(n):
    f = 1.0
    for i in range(1,2*n+2):
        f *= i
    return f

def s(n):
    i, f = 0, 0
    while (i<=n):
        f = f + (((-1)**i)*(x**(2*i)))/facto(i)
        i = i+1
    return f

plt.figure(figsize=(10, 10))
x=np.arange(-20,20.1,0.1)

plt.plot(x,np.sin(x)/x, label = 'sin(x)/x')
plt.plot(x,s(0), label = 'n=0')
plt.plot(x,s(1), label = 'n=1')
plt.plot(x,s(2), label = 'n=2')
plt.plot(x,s(3), label = 'n=3')
plt.plot(x,s(4), label = 'n=4')
plt.plot(x,s(5), label = 'n=5')
plt.plot(x,s(10), label = 'n=10')
plt.plot(x,s(15), label = 'n=15')
plt.plot(x,s(20), label = 'n=20')
plt.plot(x,s(23), label = 'n=23')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.axis([-20, 20, -1.5, 1.5])
```
